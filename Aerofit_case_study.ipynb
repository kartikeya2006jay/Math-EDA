```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/001/125/original/aerofit_treadmill.csv?1639992749")
```

```python
df=data.copy()
```

```python
df["MaritalStatus"].value_counts()
```

```python
df["Gender"].value_counts()
```

```python
df["is_married"] = df["MaritalStatus"].map({"Partnered":1,"Single":0})
df["is_male"] = df["Gender"].map({"Male":1,"Female":0})
df.head(10)
```

```python
df.columns
```

```python
col=["Age","Education","Miles"]
for x in col:
  plt.figure(figsize=(6,4))
  sns.histplot(df[x],kde=True , bins=5);
```

```python
df.columns
```

```python
col=["Age","Education","Usage","Fitness","Miles"]
for x in col:
  plt.figure(figsize=(7,5))
  sns.countplot(x=df[x]);
  plt.tight_layout()
  plt.savefig(f"{x}_countplot.pdf", dpi=300, bbox_inches="tight")
```

```python
df.columns
```

```python
col=["Age","Education","Usage","Fitness","Miles"]
for x in col:
  plt.figure(figsize=(6,4))
  sns.distplot(x=df[x]);
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Gender", y="Income", data=df)
plt.title("Income distribution by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="MaritalStatus", y="Income", data=df)
plt.title("Income distribution by MaritalStatus")
plt.show()
```

```python
sns.countplot(x="Gender", hue="is_married", data=df)
plt.title("Gender Distribution by Marital Status")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
sns.pairplot(df)
plt.tight_layout(pad=3)
```

```python
df.isnull().sum()
```

```python
df.groupby("Product").median(numeric_only=True)
```

```python
df["Income_Bin"] = pd.cut(df["Income"] , bins=5)
sns.countplot(x="Income_Bin",hue="Product", data=df)
plt.xticks(rotation=90)
plt.show()
```
